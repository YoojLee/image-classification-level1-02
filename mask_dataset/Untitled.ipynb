```python
i,brk = 0, False
im_path = "./train/images/"
for p in df.path:
    if p[-2] == "6":
        p = im_path + p
        for f_name in os.listdir(p):
            if not f_name.startswith("."):
                i += 1
                im = Image.open(p + '/' + f_name)
                np_im = np.array(im)
                gray_im = np.sum(np_im, axis = 2)
                plt.imshow(np_im[80:80+200, 80:80+220])
                # plt.imshow(gray_im[80:80+320, 80:80+220], cmap = 'gray')
                plt.show()
                if i == 10:
                    brk = True
                    break
    if brk:
        break
```
